```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('tips.csv')
data
```

```python
def impute_nan(data, variable, method):
    data[variable+'_median'] = data[variable].fillna(method)

method = data.total_bill.median() 
method
```

```python
impute_nan(data,'total_bill',method)
data
```

```python
import pandas as pd

tips_df = pd.read_csv('tips.csv')
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
tips_df['sex_encoded'] = label_encoder.fit_transform(tips_df['sex'])
tips_df['smoker_encoded'] = label_encoder.fit_transform(tips_df['smoker'])


one_hot_encoded_df = pd.get_dummies(tips_df[['day', 'time']], prefix=['day', 'time'])

ordinal_mapping = {'Sun': 0, 'Sat': 1, 'Thur': 2, 'Fri': 3}
tips_df['day_ordinal'] = tips_df['day'].map(ordinal_mapping)

print("Label Encoded Dataset:")
print(tips_df[['sex', 'sex_encoded', 'smoker', 'smoker_encoded']])

print("\nOne-Hot Encoded Dataset:")
print(one_hot_encoded_df)

print("\nOrdinal Encoded Dataset:")
print(tips_df[['day', 'day_ordinal']])
```
