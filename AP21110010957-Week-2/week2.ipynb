```python
# Q3.Write a Python program to draw a line with suitable label in the x axis, y axis and a title.
```

```python
import matplotlib.pyplot as plt


x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 6, 8, 10]
fig, ax = plt.subplots()
ax.plot(x_values, y_values, label='Line Plot')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Line Plot with Labels')
ax.legend()
plt.show()
```

```python
# Q4.Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
```

```python
import matplotlib.pyplot as plt
with open('data.txt', 'r') as file:
    lines = file.readlines()
    data = [list(map(float, line.strip().split())) for line in lines]
x_values = [row[0] for row in data]
y_values = [row[1] for row in data]
fig, ax = plt.subplots()
ax.plot(x_values, y_values, label='Line Plot')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Line Plot from Data File')
ax.legend()
plt.show()
```

```python
# Q5.Write a Python program to plot two or more lines with legends, different widths and colors.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_values = np.linspace(0, 10, 100)
y_values1 = np.sin(x_values)
y_values2 = np.cos(x_values)


fig, ax = plt.subplots()

ax.plot(x_values, y_values1, color='red', linewidth=2, label='Line 1')
ax.plot(x_values, y_values2, color='blue', linewidth=1, label='Line 2')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Multiple Lines with Legends')
ax.legend()
plt.show()
```

```python
# Q6.Write a Python program to plot quantities which have an x and y position.
```

```python
import matplotlib.pyplot as plt

x1 = [10, 3, 5, 16, 8]
y1 = [1, 5, 10, 18, 20]
x2 = [30, 43, 63, 27, 9]
y2 = [2, 6, 11, 20, 22]
fig, ax = plt.subplots()

ax.plot(x1, y1, 'b*', label='Set 1')
ax.plot(x2, y2, 'ro', label='Set 2')

ax.set_xlim(0, 70)
ax.set_ylim(0, 70)

ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Scatter Plot with Two Sets of Points')

ax.legend()
plt.show()

```

```python
# Q7. Create plot usingsubplots(),scatter, and some other methods on theAxesobject:
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
axes[0].scatter(x, y1, color='blue', marker='o', label='Scatter Plot 1')

axes[0].set_xlabel('X-axis Label')
axes[0].set_ylabel('Y-axis Label')
axes[0].set_title('Scatter Plot 1')
axes[0].legend()

axes[1].scatter(x, y2, color='green', marker='x', label='Scatter Plot 2')

axes[1].set_xlabel('X-axis Label')
axes[1].set_ylabel('Y-axis Label')
axes[1].set_title('Scatter Plot 2')
axes[1].legend()

plt.tight_layout()

plt.show()
```

```python
# Q8. Generate random data for x and y and create a scatter plot using Matplotlib.
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42) 
num_points = 50
x_values = np.random.rand(num_points) * 10 
y_values = np.random.rand(num_points) * 5  

plt.scatter(x_values, y_values, color='purple', marker='o', label='Random Data')

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

plt.title('Scatter Plot with Random Data')

plt.legend()
plt.show()
```

```python
# Q9. Create a bar chart to compare the sales of different products. You can use fictional data or come up with your own dataset.
```

```python
import matplotlib.pyplot as plt
import numpy as np

products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [120, 200, 80, 150, 100]

plt.bar(products, sales, color='skyblue')

plt.xlabel('Products')
plt.ylabel('Sales (in units)')

plt.title('Sales Comparison of Different Products')

plt.show()
```

```python
# Q10. Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of the data.
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(data, bins=30, color='lightblue', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')

plt.title('Histogram of Random Dataset')

plt.show()
```

```python
# Q11. Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.
```

```python
import matplotlib.pyplot as plt

expense_categories = ['Housing', 'Transportation', 'Food', 'Utilities', 'Entertainment']
expense_amounts = [1500, 500, 800, 300, 200]

plt.pie(expense_amounts, labels=expense_categories, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'lightyellow', 'lightsalmon'])

plt.title('Expense Distribution in Budget')
plt.show()
```
